```python
from datasets import load_dataset
import json, random, os

# ── Load dataset from HuggingFace ──
# medalpaca/medical_meadow_medqa has instruction/input/output columns
dataset = load_dataset("medalpaca/medical_meadow_medqa", split="train")
print(f"Total examples: {len(dataset)}")
print("Columns:", dataset.column_names)
print("\nSample row:")
print(json.dumps(dataset[0], indent=2))
```

```python
# ── Filter & keep only useful columns ──
# Keep rows where output is non-empty and long enough to be useful
dataset = dataset.filter(lambda x: x["output"] and len(x["output"]) > 30)
print(f"After filtering: {len(dataset)} examples")
```

```python
# ── Take 1000 examples for the assignment ──
random.seed(42)
indices = random.sample(range(len(dataset)), 1000)
dataset = dataset.select(indices)
print(f"Working dataset size: {len(dataset)}")
```

```python
# ── Category distribution (focus_area if available, else show sample) ──
print("Sample Q&A pairs:\n")
for i in range(3):
    row = dataset[i]
    print(f"--- Example {i+1} ---")
    print(f"INSTRUCTION: {row['instruction'][:120]}...")
    print(f"OUTPUT: {row['output'][:200]}...")
    print()
```

```python
# ── Save raw dataset locally ──
os.makedirs("dataset", exist_ok=True)
dataset.save_to_disk("dataset/medical_qa_raw")
print("Dataset saved to dataset/medical_qa_raw")
```
